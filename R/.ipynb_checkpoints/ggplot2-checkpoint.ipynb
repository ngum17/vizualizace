```r
# Read the dataset
df <- read.csv("healthcare_dataset2.csv", sep=";")
# View the structure of the dataset
```

# barchart

```r
start.time <- Sys.time()
# Load the ggplot2 library
library(ggplot2)

ggplot(df, aes(x = factor(Blood.Type), fill = factor(Blood.Type))) +
    geom_bar(color = "black") +
    labs(title = "Bar Chart of Blood Types", x = "Blood Type", y = "Frequency") +
    theme_minimal()

end.time <- Sys.time()

print(end.time - start.time)
```

# Histogram

```r
# Measure the execution time of a code block
start.time <- Sys.time()
# Load the ggplot2 library
library(ggplot2)
# Create a histogram of ages using ggplot2
ggplot(df, aes(x=Age)) +
  geom_histogram(breaks = seq(0, max(df$Age) + 10, by = 10), color="black", fill="skyblue") +
  labs(title="Histogram of Ages", x="Age", y="Frequency") +
  theme_minimal()
end.time <- Sys.time()

print(end.time - start.time)
```

# boxplot

```r
start.time <- Sys.time()
# Load the ggplot2 library
library(ggplot2)
ggplot(df, aes(x = factor(Medical.Condition), y = Days.in.Hospital)) +
    geom_boxplot(color = "black", fill = "skyblue") +
    labs(title = "Box Plot of Days in Hospital by Medical Condition", x = "Medical Condition", y = "Days in Hospital") +
    theme_minimal()
end.time <- Sys.time()

print(end.time - start.time)
```

```r
# Load necessary libraries

library(GGally)
library(ggplot2)
# Load the data
df <- read.csv("penguins.csv")

# Remove rows with non-finite values
df <- df[complete.cases(df), ]

# Define category orders
df$species <- factor(df$species, levels = c("Adelie", "Chinstrap", "Gentoo"))

# Create scatterplot matrix
ggpairs(df, columns = c("culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"), 
        mapping = ggplot2::aes(color = species))
```

```r
# Load necessary library
library(ggplot2)
library(gridExtra)

# Load the data
df <- read.csv("penguins.csv")
# Remove rows with non-finite values
df <- df[complete.cases(df), ]

# Define category orders
df$species <- factor(df$species, levels = c("Adelie", "Chinstrap", "Gentoo"))

# Create scatterplots
p1 <- ggplot(df, aes(x = culmen_length_mm, y = culmen_depth_mm, color = species)) + geom_point()
p2 <- ggplot(df, aes(x = culmen_length_mm, y = flipper_length_mm, color = species)) + geom_point()
p3 <- ggplot(df, aes(x = culmen_length_mm, y = body_mass_g, color = species)) + geom_point()
p4 <- ggplot(df, aes(x = culmen_depth_mm, y = flipper_length_mm, color = species)) + geom_point()
p5 <- ggplot(df, aes(x = culmen_depth_mm, y = body_mass_g, color = species)) + geom_point()
p6 <- ggplot(df, aes(x = flipper_length_mm, y = body_mass_g, color = species)) + geom_point()

# Arrange scatterplots in a grid
grid.arrange(p1, p2, p3, p4, p5, p6, ncol = 2)
```
