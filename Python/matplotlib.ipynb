# Donut chart

```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt

# declaring data 
data = [15, 30, 45, 10] 
keys = ['Frogs', 'Hogs', 'Dogs', 'Logs'] 

# define colors for the pie chart
colors = ['blue', 'orange', 'green', 'red']

# Create donut chart
fig, ax = plt.subplots()
ax.pie(data, labels = keys, colors = colors, autopct='%1.1f%%', pctdistance=0.85)

# Draw white circle in the middle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

#add legend
plt.legend(title="Pets", labels=keys, loc='center left', bbox_to_anchor=(1, 0.5)) 
#needs parameter so legend does not overlap with figure

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

plt.tight_layout()
plt.show()
```

# Scatter plot matrix

```python
%%time
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# load data stored in a CSV file
df = pd.read_csv("data/penguins.csv")

# Filter out non-numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Set the figure size
plt.figure(figsize=(12, 8))

# Create scatter plots and histograms
for i, col1 in enumerate(numeric_columns):
    for j, col2 in enumerate(numeric_columns):
        plt.subplot(len(numeric_columns), len(numeric_columns), i * len(numeric_columns) + j + 1)
        if i == j:
            plt.hist(df[col1], bins=20)  # Histogram
            plt.xlabel(col1)
            plt.ylabel('Frequency')
        else:
            plt.scatter(df[col2], df[col1], c=pd.factorize(df['species'])[0], cmap='viridis')  # Scatter plot
            plt.xlabel(col2)
            plt.ylabel(col1)
            plt.colorbar(label='species')

plt.tight_layout()
plt.show()
```

# Sankey diagram


```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# Make labels smaller
plt.rcParams['font.size'] = 5

sankey = Sankey()

# first diagram, indexed by prior=0
sankey.add(flows=[1, -1, 0.1, 0.3],
           orientations=[0, 0, 1, -1],
           labels=['input', 'output', 'top', 'bottom'],
           pathlengths=[0, 0, 0.5, 1.0])

sankey.add(flows=[1, -1],
           orientations=[0, 0],
           labels=['input2', 'output2'],
           trunklength=3.0,
           prior=0,
           connect=(1, 0))

sankey.add(flows=[1, -0.3],
           orientations=[1, -1],
           labels=['input3', 'output3'],
           prior=1,
           trunklength=2.5,
           connect=(1, 0))

sankey.finish()
```

```python
%%time
# import necessary libraries
import matplotlib as plt
import pandas as pd

testdata = pd.DataFrame({'age': [20, 20, 21, 21, 22, 22, 23, 23],
                'gender': ["male", "female", "male", "female", "male", "female", "male", "female"],
                'count': [10, 12, 13, 10, 16, 14, 17, 16]})

# Separate male and female data
male = testdata[testdata['gender'] == 'male']
female = testdata[testdata['gender'] == 'female']

# Create subplots
fig, ax = plt.subplots()

# Plot male and female data
ax.barh(male['age'], male['count'], color='blue', label='Male')
ax.barh(female['age'], -female['count'], color='orange', label='Female')

# Set x-axis labels
ax.set_xticks([-20, -10, 0, 10, 20])
ax.set_xticklabels([20, 10, 0, 10, 20])

# Add legend
ax.legend()

plt.show()
```

# Box plot

```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# load data stored in a CSV file
df = pd.read_csv('data/healthcare_dataset2.csv', sep=';')

# Create a list of colours
colors = ['red', 'green', 'blue', 'purple', 'orange', 'yellow']

# Default line colour is orange
median_props = {'color': 'black'}

# Create boxplot
box_plots = plt.boxplot([df['Days in Hospital'][df['Medical Condition'] == condition] for condition in df['Medical Condition'].unique()], 
                        labels=df['Medical Condition'].unique(), 
                        patch_artist=True, 
                        medianprops=median_props)

# For loop is used to give different colour to the box plot
for box, color in zip(box_plots['boxes'], colors):
    box.set_facecolor(color)

# Additional parameters 
plt.xlabel('Medical Condition')
plt.ylabel('Days in Hospital')
plt.title('Box Plot of Days in Hospital by Medical Condition')
plt.grid(axis = "y")
plt.show()
```

# Violin plot

```python
%%time
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# load data stored in a CSV file
df = pd.read_csv("data/penguins.csv")

# Separate data by sex
male = df[df['sex'] == 'MALE']
female = df[df['sex'] == 'FEMALE']

# Create a list of species
species = df['species'].unique()

# Create violin plots for each sex
fig, ax = plt.subplots()

for i, s in enumerate(species):
    ax.violinplot(male['flipper_length_mm'][male['species'] == s], positions=[i], widths=0.35, showmedians=True)
    ax.violinplot(female['flipper_length_mm'][female['species'] == s], positions=[i+0.4], widths=0.35, showmedians=True)
    

# Set x-axis labels
ax.set_xticks([i+0.2 for i in range(len(species))])
ax.set_xticklabels(species)

plt.title("Flipper Length for 3 Penguin Species by Sex")
plt.show()
```

# Radial bar

```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt
from matplotlib import cm
from math import log10

labels = list("ABCDEFG")
data = [21, 57, 88, 14, 76, 91, 26]
#number of data points
n = len(data)
#find max value for full ring
k = 10 ** int(log10(max(data)))
m = k * (1 + max(data) // k)

#radius of donut chart
r = 1.5
#calculate width of each ring
w = r / n 

#create colors along a chosen colormap
colors = [cm.terrain(i / n) for i in range(n)]

#create figure, axis
fig, ax = plt.subplots()
ax.axis("equal")

#create rings of donut chart
for i in range(n):
    #hide labels in segments with textprops: alpha = 0 - transparent, alpha = 1 - visible
    innerring, _ = ax.pie([m - data[i], data[i]], radius = r - i * w, startangle = 90, labels = ["", labels[i]], labeldistance = 1 - 1 / (1.5 * (n - i)), textprops = {"alpha": 0}, colors = ["white", colors[i]])
    plt.setp(innerring, width = w, edgecolor = "white")

plt.legend()
plt.show()
```

# Radar chart

```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
 
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D'],
'var1': [38, 1.5, 30, 4],
'var2': [29, 10, 9, 34],
'var3': [8, 39, 23, 24],
'var4': [7, 31, 33, 14],
'var5': [28, 15, 32, 14]
})
 
# ------- PART 1: Create background
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)
 

# ------- PART 2: Add plots
 

# Ind1
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="group A")
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=df.loc[1].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="group B")
ax.fill(angles, values, 'r', alpha=0.1)
 
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the graph
plt.show()
```

# Gantt chart

```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define data
df = pd.DataFrame([
    dict(Task="Job A", Start='2009-01-01', Finish='2009-02-28'),
    dict(Task="Job B", Start='2009-03-05', Finish='2009-04-15'),
    dict(Task="Job C", Start='2009-02-20', Finish='2009-05-30')
])

# Convert dates to datetime
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])

# Convert datetime to numeric for plotting
df['Start_num'] = pd.to_numeric(df['Start'])
df['Finish_num'] = pd.to_numeric(df['Finish'])

# Create figure and axes
fig, ax = plt.subplots()

# Loop over tasks
for i, task in enumerate(df['Task']):
    # Calculate start and duration
    start = df.loc[df['Task'] == task, 'Start_num'].values[0]
    finish = df.loc[df['Task'] == task, 'Finish_num'].values[0]
    duration = finish - start

    # Add a broken_barh for each task
    ax.broken_barh([(start, duration)], (i-0.4, 0.8))

# Set yticks and labels
ax.set_yticks(np.arange(len(df['Task'])))
ax.set_yticklabels(df['Task'])

# Set xticks and labels
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: pd.to_datetime(x).strftime('%Y-%m-%d')))

# Invert y-axis
ax.invert_yaxis()

# Show the plot
plt.show()
```

# Heatmap

```python
%%time
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data stored in a CSV file
df = pd.read_csv("data/penguins.csv")

# Remove NaN values
df = df.dropna()

# Extract numeric columns for heatmap
numeric_data = df.select_dtypes(include=['float64'])

# Create a correlation matrix
corr_matrix = numeric_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='viridis', interpolation='nearest')

# Add color bar
plt.colorbar()

# Add values of correlation
for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}',
                 ha='center', va='center', color='black')

# Set labels
plt.title('Correlation Heatmap of Penguin Dataset')
plt.xticks(ticks=np.arange(len(corr_matrix.columns)), labels=corr_matrix.columns)
plt.yticks(ticks=np.arange(len(corr_matrix.index)), labels=corr_matrix.index)

# Show plot
plt.show()
```

# Bubble chart

```python
%%time
# import necessary libraries
import matplotlib.pyplot as plt

# Data
browser_market_share = {
    'browsers': ['firefox', 'chrome', 'safari', 'edge', 'ie', 'opera'],
    'market_share': [8.61, 69.55, 8.36, 4.12, 2.76, 2.43],
    'color': ['#5A69AF', '#579E65', '#F9C784', '#FC944A', '#F24C00', '#00B825']
}

# Create a bubble chart
fig, ax = plt.subplots()

# Scale factor for the size of the bubbles
scale_factor = 100

# Using a scatter plot to create the bubble chart
scatter = ax.scatter(browser_market_share['browsers'], 
                     browser_market_share['market_share'],
                     s=[ms * scale_factor for ms in browser_market_share['market_share']], 
                     c=browser_market_share['color'], 
                     alpha=1, 
                     edgecolors="w", 
                     linewidth=2)

# Adding title and labels
ax.set_title('Browser market share')
ax.set_xlabel('Browser')
ax.set_ylabel('Market Share')

plt.show()
```
