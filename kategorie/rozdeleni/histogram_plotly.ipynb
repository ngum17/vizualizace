```python
%%time
# import necessary libraries
import plotly.express as px
import pandas as pd

# load data stored in a CSV file
df = pd.read_csv('data/healthcare_dataset2.csv', sep=';')

# Create a histogram of ages
fig = px.histogram(df, x='Age', nbins=10, title='Histogram of Ages', 
                   labels={'Age': 'Age', 'count': 'Frequency'},color_discrete_sequence=['skyblue'])



# Set the size of the plot to 300x300
fig.update_layout(autosize=False, width=400, height=400)

# Show the plot
fig.show()
```
